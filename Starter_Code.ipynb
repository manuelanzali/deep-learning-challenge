## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN'])
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_type_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_counts
```

```python
application_df.loc[application_df['INCOME_AMT'] == '0', 'INCOME_AMT'] = 0
application_df.loc[application_df['INCOME_AMT'] == '1-9999', 'INCOME_AMT'] = 1
application_df.loc[application_df['INCOME_AMT'] == '10000-24999', 'INCOME_AMT'] = 2
application_df.loc[application_df['INCOME_AMT'] == '25000-99999', 'INCOME_AMT'] = 3
application_df.loc[application_df['INCOME_AMT'] == '100000-499999', 'INCOME_AMT'] = 4
application_df.loc[application_df['INCOME_AMT'] == '1M-5M', 'INCOME_AMT'] = 5
application_df.loc[application_df['INCOME_AMT'] == '5M-10M', 'INCOME_AMT'] = 6
application_df.loc[application_df['INCOME_AMT'] == '10M-50M', 'INCOME_AMT'] = 7
application_df.loc[application_df['INCOME_AMT'] == '50M+', 'INCOME_AMT'] = 8
application_df['INCOME_AMT'] = application_df['INCOME_AMT'].astype(int)
application_df['INCOME_AMT'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(application_type_counts[application_type_counts < 400].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
CLASSIFICATION_counts = application_df['CLASSIFICATION'].value_counts()
CLASSIFICATION_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
CLASSIFICATION = CLASSIFICATION_counts[CLASSIFICATION_counts > 1]
CLASSIFICATION
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(CLASSIFICATION_counts[CLASSIFICATION_counts < 2000].index)

    #classifications_to_keep = [1000, 1200, 2000, 2100, 3000]

#Create a mask for classifications to keep


    #mask = application_df['CLASSIFICATION'].isin(classifications_to_keep)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

    #application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].where(mask, "Other")


# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
    #print(application_df['CLASSIFICATION'].unique())
```

```python
#Look at NAME value counts to identify and replace with "Other"
NAME_counts = application_df['NAME'].value_counts()
NAME_counts
```

```python
# Choose a cutoff value and create a list of name types to be replaced
# use the variable name `NAME_to_replace`
NAME_to_replace = list(NAME_counts[NAME_counts < 50].index)

# Replace in dataframe
for nam in NAME_to_replace:
    application_df['NAME'] = application_df['NAME'].replace(nam,"Other")

# Check to make sure replacement was successful
application_df['NAME'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
application_df_dummies = pd.get_dummies(application_df)
application_df_dummies
```

```python
# Split our preprocessed data into our features and target arrays
X = application_df_dummies.drop(columns = 'IS_SUCCESSFUL')
y = application_df_dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features = len(X_train_scaled[0])
hidden_nodes_layer1 = num_input_features
hidden_nodes_layer2 = round(hidden_nodes_layer1/2)
hidden_nodes_layer3 = round(hidden_nodes_layer2/2)

nn = tf.keras.models.Sequential()

# Input layer
nn.add(tf.keras.layers.Input(shape=(num_input_features,)))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, activation="sigmoid", input_dim=num_input_features))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="sigmoid"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="sigmoid"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
my_fit_model = nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
